#### 8. What are the top 5 most populated States according to the 2010 Census?

```python
# group by State - all columns
# sort by 2010Census, asc
# reset index to make State as column and pick the top 5 states

pop.groupby(['State']).sum().sort_values(by='2010Census', ascending=False).reset_index()[['State']][:5]

# pop[['State','2010Census']].groupby('State').sum().sort_values('2010Census', ascending=False).head().index.to_list()
```

#### 11. Add a column that calculates the percent change between the 2010 Census and the 2017 Population Estimate

```python
# Write your code here

pop['PerChangebyCounty'] = ((pop['2017PopEstimate']-pop['2010Census'])*100/pop['2010Census']).round(2)
pop.head(10)
```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.

```python
# Write your code here

# group by State, take sum of cols = 2010Census and 2017PopEstimate
pop_copy = pop.groupby(['State'])[['2010Census', '2017PopEstimate']].sum().reset_index()
# recalculate PercentChange by State
pop_copy['PerChangebyState'] = ((pop_copy['2017PopEstimate']-pop_copy['2010Census'])*100/pop_copy['2010Census']).round(2)
# sort by PercentChange desc and pick top 5 states
pop_copy.sort_values(by='PerChangebyState', ascending=False).reset_index(drop=True)[:5]
```
